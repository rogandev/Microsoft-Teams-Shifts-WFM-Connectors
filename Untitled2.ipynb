```python
# prompt: Using the tensorboard embedding projector. prompt for a csv file, take the uploaded csv file and create an Keras embedding layer from the uploaded file using 16 as the embedding dimension, then configure the embedding layer as part of a keras model, compile model, train model for one epoch, save the data for tensorboard as tsv file

import io
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Embedding
from keras.optimizers import Adam
from google.colab import files
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorboard.plugins import projector

uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{0}" with length {1} bytes'.format(fn, len(uploaded[fn])))

data = pd.read_csv(io.BytesIO(uploaded[fn]))
print(data)

# Create an embedding layer
embedding_dim = 16
embedding_layer = Embedding(data.shape[0], embedding_dim)

# Create a model
#model = Sequential()
#model.add(embedding_layer)
#model.add(Adam(learning_rate=0.001))
model = tf.keras.Sequential(
    [
        embedding_layer, # The embedding layer should be the first layer in a model.
        tf.keras.layers.GlobalAveragePooling1D(),
        tf.keras.layers.Dense(16, activation="relu"),
        tf.keras.layers.Dense(1),
    ]
)
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(tf.data.Dataset.from_tensor_slices((data.values.tolist(), data.values.tolist())), epochs=1)

# Save the data for tensorboard
model.save('embedding_model.h5')

# Export the embedding weights to a TSV file
embedding_weights = model.layers[0].get_weights()[0]
np.savetxt('embedding_weights.tsv', embedding_weights, delimiter='\t')
```
